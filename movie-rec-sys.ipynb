```python
import pandas as pd

movies = pd.read_csv("tmdb_5000_movies.csv", low_memory=False)
credits = pd.read_csv("tmdb_5000_credits.csv", low_memory=False)

print(movies.shape)
print(credits.shape)
```

```python
movies.sample(5)
```

```python
credits.sample(5)
```

```python
credits = credits.rename(columns={'movie_id': 'id'})

df = movies.merge(credits, on='id')
```

```python
df.shape
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
# select only needed columns
df = df[['id','title_x','genres','keywords','overview','cast','crew']]

df = df.copy()

# fill missing overview
df.loc[:, 'overview'] = df['overview'].fillna('')
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
df.info()
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
df['genres'] = df['genres'].apply(convert)
df['keywords'] = df['keywords'].apply(convert)
```

```python
df['cast'][0]
```

```python
def convert_cast(text):
    casts = []

    for i in ast.literal_eval(text)[:3]:
        casts.append(i['name'])

    return casts
```

```python
df['cast'] = df['cast'].apply(convert_cast)
```

```python
df['cast'][0]
```

```python
df['crew'][0]
```

```python
def fetch_director(text):
    directors = []

    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            directors.append(i['name'])

    return directors
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df["crew"][0]
```

```python
def remove_space(L):
    return [i.replace(" ", "") for i in L]

df['genres'] = df['genres'].apply(remove_space)
df['keywords'] = df['keywords'].apply(remove_space)
df['cast'] = df['cast'].apply(remove_space)
df['crew'] = df['crew'].apply(remove_space)
```

```python
df['overview'] = df['overview'].apply(lambda x: x.lower().split())
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
new_df = df[['id','title_x','tags']].copy()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x).lower())
```

```python
print(new_df['tags'][0])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
vectors = cv.fit_transform(new_df['tags']).toarray()

vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity.shape
```

```python
def recommend(movie):
    index = new_df[new_df['title_x'] == movie].index[0]
    distances = similarity[index]
    movies_list = list(enumerate(distances))
    movies_list = sorted(movies_list, key=lambda x: x[1], reverse=True)[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title_x)
```

```python
recommend("Avatar")
```

```python
recommend("Gandhi")
```

```python
import pickle

pickle.dump(new_df, open('movies.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
pickle.dump(cv, open('vectorizer.pkl', 'wb'))
```
